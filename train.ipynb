```python
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import KFold
from scipy import sparse
import random
```

```python
train1 = np.loadtxt('/home/kesci/data/V6_train_data__6_10.txt')
y1 = np.loadtxt('/home/kesci/data/V6_labels__6_10.txt')
train2 = np.loadtxt('/home/kesci/data/V6_train_data_1_17.txt')
y2 = np.loadtxt('/home/kesci/data/V6_labels_1_17.txt')
train3 = np.loadtxt('/home/kesci/data/V6_train_data_8_24.txt')
y3 = np.loadtxt('/home/kesci/data/V6_labels_8_24.txt')

train = np.vstack((train1,train2,train3))
y = np.hstack((y1,y2,y3))

print(train.shape)
test = np.loadtxt('/home/kesci/data/V6_test_data.txt')#sparse.load_npz('test_0715.npz').toarray()
userId = np.loadtxt('/home/kesci/data/V6_userId.txt',dtype=str)
kf = KFold(n_splits=5, shuffle=True, random_state=1024)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=1024)
feature_importance = 0
test_y = np.zeros(y.shape)
pred_y = np.zeros((test.shape[0],))
for train_index,val_index in kf.split(train):
    X_train,X_val = train[train_index],train[val_index]
    y_train,y_val = y[train_index],y[val_index]
    clf = lgb.LGBMClassifier(boosting_type='dart',learning_rate=0.005,objective='binary',reg_alpha=0,
                             subsample=0.8,colsample_bytree=0.8,n_estimators=200000,
                             early_stopping_round=500,silent=-1)
    clf.fit(X_train,y_train,eval_set=[(X_val,y_val)],eval_metric='auc',verbose=100,early_stopping_rounds=500)
    feature_importance += clf.feature_importances_
    test_y[val_index] = clf.predict_proba(X_val,num_iteration=clf.best_iteration_)[:,1]
    pred_y += clf.predict_proba(test,num_iteration=clf.best_iteration_)[:,1]
```

```python
from catboost import CatBoostClassifier
kf = KFold(n_splits=5, shuffle=True, random_state=1024)

test_y = np.zeros(y.shape)
pred_y = np.zeros((test.shape[0],))
for train_index,val_index in kf.split(train):
    X_train,X_val = train[train_index],train[val_index]
    y_train,y_val = y[train_index],y[val_index]
    clf = CatBoostClassifier(iterations=12000,learning_rate=0.01,loss_function='Logloss',
                             custom_metric='AUC',od_type='Iter',od_wait=300,
                             eval_metric='AUC',verbose=100,thread_count=8,
                             bootstrap_type='Bernoulli',subsample=0.8,rsm=0.7)
    clf.fit(X_train,y_train,eval_set=[(X_val,y_val)],verbose=100)
    #feature_importance += clf.feature_importances_
    test_y[val_index] = clf.predict_proba(X_val)[:,1]
    pred_y += clf.predict_proba(test)[:,1]
```

```python
D = {}
for i in range(len(userId)):
    D[userId[i]] = pred_y[i]/5.0
with open('/home/kesci/work/LGB-6with1with8_NEW.txt','w') as f:
    for i in D:
        f.write(i+','+str(D[i])+'\n')
```

```python
import xgboost as xgb

test_y = np.zeros(y.shape)
pred_y = np.zeros((test.shape[0],))
for train_index,val_index in kf.split(train):
    X_train,X_val = train[train_index],train[val_index]
    y_train,y_val = y[train_index],y[val_index]
    DTrain = xgb.DMatrix(X_train,label=y_train,silent=True)
    DVal = xgb.DMatrix(X_val,label=y_val,silent=True)
    DTest = xgb.DMatrix(test,silent=True)
    param = {'eta':0.02,'n_estimators':20000,'objective':'rank:pairwise',
             'eval_metric':'auc','silent':True,'subsample':0.8,'colsample_bytree':0.8,
             'reg_alpha':0.002
            }
    clf = xgb.train(param,DTrain,num_boost_round=20000,evals=[(DVal,'validation')],
                    early_stopping_rounds=300,verbose_eval=100)
    test_y[val_index] = clf.predict(DVal)
    pred_y += clf.predict(DTest)
```

```python
from sklearn import cross_validation,metrics
test_auc = metrics.roc_auc_score(y,test_y)#auc
print('auc:'+str(test_auc))
```
